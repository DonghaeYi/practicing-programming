```python
#singly linked list

class sLL:
    #default constructor
    def __init__(self):
        self.head = None
        self.length = 0

    def add_node(self,new_data):
        new_node = Node()
        new_node.set_data(new_data) 
        temp = self.head
        
        if temp == None:
            self.head = new_node
            self.head.next = None
            self.length += 1
            return
            
        while temp.next != None:
            temp = temp.next
        temp.next = new_node
        new_node.next = None
        self.length += 1
    
    def search_node(self, search_data):
        temp = self.head
        while temp != None:
            if temp.data == search_data:
                return temp
            temp = temp.next
        return None
            
    def print_ll(self):
        temp = self.head
        while temp != None:
            print(f"{temp.data} ---> ", end = "")
            temp = temp.next
        print("None")

    def delete_idx(self,index):
        prev = None
        if index < 0 or index >= self.length:
        	return
        if index == 0:
            self.head = self.head.next
        else:
            curr = self.head 
            while index != 0:
                prev = curr
                curr = curr.next
                index -= 1
            prev.next = curr.next
            curr = None
        self.length -= 1
    
            
        
            
class Node:
    def __init__(self):
        self.data = 0
        self.next = None
        
    def get_data(self):
        return self.data
    
    def set_data(self, new_data):
        self.data = new_data
    
    def get_next(self):
        return self.next
    
    def set_next(self, new_next):
        self.next = new_next



```

```python
arr = [5, 15, 20, 25]
test_llist = sLL()
for i in arr:
    test_llist.add_node(i)
test_llist.print_ll()
test_llist.delete_idx(0)
test_llist.print_ll()
test_llist.delete_idx(1)
test_llist.print_ll()
test_llist.delete_idx(10)
test_llist.print_ll()
```

```python
class queue:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def push_q(self, new_data):
        new_node = node()
        new_node.set_data(new_data)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
            self.length += 1
            return
        self.tail.next = new_node
        self.tail = new_node
        self.length += 1
        
    def pop_q(self):
        if self.length == 0:
            return
        temp = self.head
        self.head = temp.next
        # delete temp
        temp = None
        self.length -= 1

    def print_q(self):
        temp = self.head
        while temp != None:
            print(f"{temp.data}--->", end = "")
            temp = temp.next
        print("None")

    def search_q(self, data):
        temp = self.head
        while temp != None:
            if temp.data == data:
                return temp
            temp = temp.next
        print("No data found")


class node:
    def __init__(self):
        self.data = 0
        self.next = None

    def get_data(self):
        return self.data

    def set_data(self, new_data):
        self.data = new_data

    def get_next(self):
        return self.next

    def set_next(self, new_next):
        self.next = new_next
```
